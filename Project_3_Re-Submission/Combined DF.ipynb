```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
```

```python
#Imports NYC DF - DC/EDA
NYCDF = pd.read_csv('NYCDF.csv')
```

```python
#Imports Boston DF - DC/EDA
BostonDF = pd.read_csv('BostonDF.csv')
```

```python
#Combines Boston and NYC DF - PPM
DFBoth = pd.concat([NYCDF, BostonDF], axis=0)
```

```python
#Confirms DF created - PPM
DFBoth
```

```python
#Gives each Subreddit a value of 1 or 0 depending on BOS or NYC - PPM
DFBoth['Subreddit'] = DFBoth['Subreddit'].map({'boston': 0, 'nyc': 1})
DFBoth.head()
```

```python
#Gets value counts of dataframes - PPM
DFBoth['Subreddit'].value_counts(normalize=True)
```

```python
#Gets number of Boston vs NYC subreddits - PPM
(DFBoth['Subreddit'] == 0).sum(), (DFBoth['Subreddit'] == 1).sum()
```

```python
def basic_model(title_obs):
    title = str(title_obs)
    if 'nyc' in title.lower():
        return 1
    else:
        return 0
```

```python
#Basic Model to build off of - PPM
basic_model("I love pizza in NYC!"), basic_model("I love pizza in boston!")
```

```python
#Set predictions - PPM
y_pred = [basic_model(title) for title in list(DFBoth['Title'])]
```

```python
#Sets actual - PPM
y_actual = list(DFBoth['Subreddit'])
y_actual[:5]
```

```python
#Sets total correct - PPM
total_correct = sum([1 if _pred == _actual else 0 
                    for _pred, _actual in zip(y_pred, y_actual)]
                     )
total_correct
```

```python
#Sets total observations - PPM
total_observations = len(y_actual)
total_observations
```

```python
#Gets accuracy - PPM
accuracy = total_correct / total_observations
accuracy
```
