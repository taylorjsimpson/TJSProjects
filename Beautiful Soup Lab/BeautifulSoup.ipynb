### Step 1: Create a soup object from the home page

````python
# You'll need the requests library in order to fully utilize bs4.
import requests
from bs4 import BeautifulSoup as soup

# Target web page:
url = "http://chubbygrub.com"

# Establishing the connection to the web page:
response = requests.get(url)

soup = soup(response.content, 'lxml')
print(soup.prettify())

# You can use status codes to understand how the target server responds to your request.
# Ex., 200 = OK, 400 = Bad Request, 403 = Forbidden, 404 = Not Found.
#print('Status Code: ',response.status_code)

# Pull the HTML string out of requests and convert it to a Python string.
html = response.text

# The first 700 characters of the content.
#print("\nFirst part of HTML document fetched as string:\n")
#print(html[:12000])
````

````python
#find all of class restaurant buttons
div_results = soup.body.find_all('div', {'class':'restaurant-buttons'})
# div_results
restaurants=[]
for restaurant_link in div_results[0].find_all('a'):
#     print(type(restaurant_link),' is ',restaurant_link)
    restaurant={}
    restaurant['slug']=restaurant_link['href']
    restaurant['name']=restaurant_link.text
    restaurants.append(restaurant)
restaurants
````
